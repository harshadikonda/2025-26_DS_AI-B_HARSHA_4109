```python
# create tuple
t = (1, 2, 3, 4, 2)

print(t[0])
print(t[-1])
print(t[1:4])

print(len(t))
print(t.count(2))
print(t.index(3))

print(2 in t)

t2 = (5, 6)
print(t + t2)
print(t * 2)

for i in t:
    print(i, end=" ")

a, b, c, d, e = t
print("\nUnpacked:", a, b, c, d, e)

lst = list(t)
tpl = tuple(lst)
```
